```python
#Load the test data
test_data = pd.read_csv("../input/testdata/test.csv")
test_data.head()
```

```python
#check for missing values
train_data.isna().sum()
```

```python
#filling missing values in train data set
train_data['Age'].fillna(train_data['Age'].mean(),inplace= True)
train_data.drop(columns = ["Cabin"], axis = 1, inplace = True)
train_data.head()
```

```python
#filling missing values in test data set
test_data['Age'].fillna(test_data['Age'].mean(),inplace= True)
test_data.drop(columns = ["Cabin"], axis = 1, inplace = True)
test_data.head()
```

```python
#Calculate Size and is_Alone columns for train data set
train_data['Size'] = train_data['SibSp'] + train_data['Parch']
train_data['is_Alone'] = [0 if x<1 else 1 for x in train_data['Size']]
train_data.head()
```

```python
#Calculate Size and is_Alone columns for test data set
test_data['Size'] = test_data['SibSp'] + test_data['Parch']
test_data['is_Alone'] = [0 if x<1 else 1 for x in test_data['Size']]
test_data.head()
```

```python
#Calculate Fare range
train_data['Fare_range'] = pd.qcut(train_data['Fare'], 5)
train_data[["Fare_range","Survived"]].groupby(["Fare_range"],as_index = False).mean().sort_values(by = "Survived",ascending = False)
```

```python
#Calculate Fare range
train_data['Fare_range'] = [0 if 39.688<x<512.329 else 1 if 21.679<x<39.688 else 2 if 10.5<x<21.679 else 4 if 7.854<x<10.5 else 4 for x in train_data['Fare']]
train_data.head()
```

```python
#Calculate Fare range
test_data['Fare_range'] = [0 if 39.688<x<512.329 else 1 if 21.679<x<39.688 else 2 if 10.5<x<21.679 else 4 if 7.854<x<10.5 else 4 for x in test_data['Fare']]
test_data.head()
```

```python
#Calculate Age range
train_data['Age_range'] = pd.qcut(train_data['Age'], 5)
train_data[["Age_range","Survived"]].groupby(["Age_range"],as_index = False).mean().sort_values(by = "Age_range",ascending = False)
```

```python
#Calculate Age range
train_data['Age_range'] = [0 if 38.0<x<80.0 else 1 if 29.699<x<38.0 else 2 if 28.0<x<29.699 else 4 if 20.0<x<28.0 else 4 for x in train_data['Age']]
train_data.head()
```

```python
#Calculate Age range
test_data['Age_range'] = [0 if 38.0<x<80.0 else 1 if 29.699<x<38.0 else 2 if 28.0<x<29.699 else 4 if 20.0<x<28.0 else 4 for x in test_data['Age']]
test_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch", "Age", "Embarked","is_Alone"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
